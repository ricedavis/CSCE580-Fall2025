# a) Number of classes and their dates

```python
# Count number of unique classes
num_classes = data["class"].nunique()

# Get class-date pairs
class_dates = data[["class", "date"]].drop_duplicates().sort_values("class")
num_classes, class_dates
```

# b) Median Class attendance per class

```python
attendance_per_class = data.groupby(["class", "date"]).size().reset_index(name="attendance")

print("Median attendance for all classes: ", attendance_per_class['attendance'].median(), "students")
```

# c) Dates with lowest and highest attendance

```python
min_att = attendance_per_class["attendance"].min()
max_att = attendance_per_class["attendance"].max()

lowest = attendance_per_class[attendance_per_class["attendance"] == min_att]
highest = attendance_per_class[attendance_per_class["attendance"] == max_att]

print('Lowest attendance:')
print(lowest)
print('Highest attendance:')
print(highest)
```

# d) Correlation With Course Evaluations & When Attendance Is Highest

```python
evaluation_classes = [15, 24, 26]  # course evaluation dates (class 15, 24 and 26)

attendance_per_class["is_evaluation"] = attendance_per_class["class"].isin(evaluation_classes).astype(int)
```

```python
eval_avg = attendance_per_class[attendance_per_class["is_evaluation"] == 1]["attendance"].mean()
non_eval_avg = attendance_per_class[attendance_per_class["is_evaluation"] == 0]["attendance"].mean()

correlation = attendance_per_class["attendance"].corr(attendance_per_class["is_evaluation"])

eval_avg, non_eval_avg, correlation
```
